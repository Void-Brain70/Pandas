```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/weather_data.csv")
df
```

```python
new_df = df.replace(-99999,np.NaN)
new_df
```

```python
new_df = df.replace([-99999,-88888,0],np.NaN)
new_df
```

```python
new_df = df.replace({
    'temperature':-99999,
    'windspeed':-99999,
    'event': '0'
},np.NaN)
new_df
```

```python
new_df = df.replace({
    -99999:np.NaN,
    '0':"Sunny"
})
new_df
```

```python
new_df = df.replace('[A-Z,a-z]','',regex=True)
new_df
```

```python
new_df = df.replace({
    'temperature':'[A-Za-z]',
     'windspeed':'[A-Za-z]'
    },'',regex=True)
new_df
```

```python
df = pd.DataFrame({
    'marks':["92","85","75","77","82"],
    "Name":["Anik","Rab","Abir","Obi","Oni"]
})
df
```

```python
new_df = df.replace(["92","85","75","77","82"],["A+","A+","A","A","A+"])
new_df
```
